## Decting how many no of outliers

   * IQR (interqurantile range) is used to detect the outliers in all numerical columns.


# Steps :

    * calculate the Q1(25th percentile) and Q3(75th percentile).
    * find IQR=Q3-Q1
    * define lower bound and upper bound
    * lower bound : q1-1.5*IQR
    * upper bound : q3+1.5*IQR
    * any value outside this range is considered as outliers

```python
drug.describe() # to describe the numerical column for find outliers
```

# observation

   * it shows the numerical column for find the outliers

```python
# to find Q1 and Q3 values

Q1 = drug['Drug_Experimentation'].quantile(0.25)
Q3 = drug['Drug_Experimentation'].quantile(0.75)
display(Q1)
display(Q3)
```

```python
# to find IQR

IQR = Q3-Q1 # it will be synatx for find IQR
display(IQR)
```

```python
# to find lower and upper limit to set boundries

lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
display(lower_limit)
display(upper_limit)
```

```python
# to find outliers

outliers = drug[(drug['Drug_Experimentation']<lower_limit) | (drug['Drug_Experimentation']>upper_limit)]
display(outliers)
```

```python
#to show the boxplot for the outliers

sns.boxplot(x=drug['Drug_Experimentation']) # syntax for seaborn boxplot
plt.title('boxplot for drug experimentation') # title for the boxplot
plt.show()
```

# observation

  * the most of the drug experimentation are in (25 - 45 years)

### 3. Which socioeconomic status group has the highest average drug experimentation?

```python
avg_drug=drug.groupby('Socioeconomic_Status')['Drug_Experimentation'].mean()
display(avg_drug)
```

```python
# Avg drug experimentation
plt.figure(figsize=(8, 6))
plt.bar(avg_drug.index, avg_drug.values, color=['purple', 'orange', 'grey'])
plt.title('Average Drug Experimentation by Socioeconomic Status')
plt.xlabel('Socioeconomic Status')
plt.ylabel('Average Drug Experimentation')
plt.show()
```

# observation

   * The lowest level drug experimentaion is 40.331692

   * The highestlevel drug experimentaion is 40.146790

   * The middle level  drug experimentaion is 39.974907

### 4. How does mental health score relate to parental supervision?

### 5. What is the trend of smoking prevalence and drug experimentation over the years?

```python
trends = drug.groupby('Year')[['Smoking_Prevalence', 'Drug_Experimentation']].mean()
display(trends)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(trends.index, trends['Smoking_Prevalence'], marker='o', label='Smoking Prevalence', color='red')
plt.plot(trends.index, trends['Drug_Experimentation'], marker='x', label='Drug Experimentation', color='blue')
plt.title('Trends of Smoking Prevalence & Drug Experimentation Over Years')
plt.xlabel('Year')
plt.ylabel('Average Percentage')
plt.legend()
plt.grid(True)
plt.show()
```
